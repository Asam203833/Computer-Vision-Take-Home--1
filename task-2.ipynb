# Image smoothing with mean filters
This notebook loads an image named **task-2**, converts it to grayscale, and applies mean filters of sizes 3 × 3, 10 × 10, and 20 × 20.

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from scipy.ndimage import uniform_filter

# -------------------------------------------
# 1. Load the image 
# -------------------------------------------
img_name = 'task-2.jpg'  
img = Image.open(img_name)
img_arr = np.array(img, dtype=np.float32)

# -------------------------------------------
# 2. Define neighbourhood sizes & process
# -------------------------------------------
neighborhoods = [3, 10, 20]
results = {}

for size in neighborhoods:
    # uniform_filter performs an arithmetic mean over an N×N window
    filtered = uniform_filter(img_arr, size=size, mode='constant', cval=0)
    results[size] = filtered.astype(np.uint8)

# -------------------------------------------
# 3. Display and save the outputs
# -------------------------------------------
plt.figure(figsize=(15, 5))
for i, (size, filt_img) in enumerate(results.items()):
    plt.subplot(1, 3, i + 1)
    plt.imshow(filt_img, cmap='gray')
    plt.title(f"{size}×{size} mean filter")
    plt.axis('off')
plt.show()
print('Done. Images saved as task-2_3x3.png, task-2_10x10.png, task-2_20x20.png')
```
