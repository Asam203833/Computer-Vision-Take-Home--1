# Image rotation demo
This notebook loads **task-3.jpg**, rotates it by 45 ° and 90 °, displays the results, and saves the two rotated images.

```python
# If Pillow or Matplotlib are missing, uncomment the next line
# !pip install pillow matplotlib --quiet
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import pathlib

# --------------------------------------------------
# 1. Load the image
# --------------------------------------------------
img_path = pathlib.Path('task-3.jpg')
if not img_path.exists():
    raise FileNotFoundError('task-3.jpg not found in this directory')

img = Image.open(img_path)
print(f'Loaded {img_path.name} → size: {img.size}, mode: {img.mode}')
```

```python
# --------------------------------------------------
# 2. Perform the rotations
# --------------------------------------------------
angles = [45, 90]
rotated_images = {}

for angle in angles:
    # expand=True enlarges the canvas so the whole image fits after rotation
    rotated = img.rotate(angle, expand=True)
    rotated_images[angle] = rotated
```

```python
# --------------------------------------------------
# 3. Display original and rotated versions side-by-side
# --------------------------------------------------
plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Original')
plt.axis('off')

for i, angle in enumerate(angles, start=2):
    plt.subplot(1, 3, i)
    plt.imshow(rotated_images[angle])
    plt.title(f'Rotated {angle}°')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
